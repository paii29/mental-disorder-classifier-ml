```python
mhds['Suicidal thoughts'] = mhds['Suicidal thoughts'].replace('YES ', "YES")
mhds['Suicidal thoughts'].value_counts()
```

```python
mhds = mhds.drop('Patient Number', axis =1)
mhds.head()
```

```python

mhds['Expert Diagnose'] = mhds['Expert Diagnose'].map({
    'Normal': 0, 
    'Bipolar Type-1': 1,
    'Bipolar Type-2': 2,
    'Depression': 3,
}).astype(int)
mhds.head()
```

```python
yes_no_cols = ['Mood Swing', 'Suicidal thoughts', 'Anorxia', 'Authority Respect', 'Try-Explanation', 'Aggressive Response', 'Ignore & Move-On', 'Nervous Break-down', 'Admit Mistakes', 'Overthinking']
for col in yes_no_cols:
    mhds[col] = mhds[col].map({'YES': 1, 'NO': 0}).astype(int)
```

```python
mhds.head()
```

```python
tonum_cols = ['Sexual Activity', 'Concentration', 'Optimisim']

for col in tonum_cols:
    mhds[col] = mhds[col].map({'0 From 10': 0, '1 From 10': 1, '2 From 10': 2, '3 From 10': 3, '4 From 10': 4, '5 From 10': 5,'6 From 10': 6,'7 From 10': 7, '8 From 10': 8,'9 From 10': 9, '10 From 10': 10}).astype(int)
```

```python
mhds.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the column containing string labels
mhds['Sadness'] = label_encoder.fit_transform(mhds['Sadness'])
mhds['Euphoric'] = label_encoder.fit_transform(mhds['Euphoric'])
mhds['Exhausted'] = label_encoder.fit_transform(mhds['Exhausted'])
mhds['Sleep dissorder'] = label_encoder.fit_transform(mhds['Sleep dissorder'])
```

```python
mhds.head()
```

```python
mhds['Sexual Activity'] = label_encoder.fit_transform(mhds['Sexual Activity'])
mhds['Concentration'] = label_encoder.fit_transform(mhds['Concentration'])
mhds['Optimisim'] = label_encoder.fit_transform(mhds['Optimisim'])
```

```python
mhds.head()
```

```python
X = mhds.drop('Expert Diagnose',axis=1) #features
y = mhds['Expert Diagnose'] #target 
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train_encoded, y_train)
```

```python
y_pred_logreg = logreg.predict(X_test_encoded)
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
```

```python
accuracy_logreg
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred_logreg))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_logreg)
```
